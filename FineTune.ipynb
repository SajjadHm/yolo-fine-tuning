```python
def txt_to_df(Vid_number):
    f = open(f'/content/drive/MyDrive/Synopsis dataset/Vid{Vid_number}_Annotation.txt' , 'r')
    frame_numbers = []
    Xs = []
    Ys = []
    Ws = []
    Hs = []
    for line in f.readlines():
        splited_line = line.split()
        frame_numbers.append(splited_line[2])
        Xs.append(splited_line[3])
        Ys.append(splited_line[4])
        Ws.append(splited_line[5])
        Hs.append(splited_line[6])

    df = pd.DataFrame( data = frame_numbers , columns=['frame_number'] )
    df['X'] = Xs
    df['Y'] = Ys
    df['W'] = Ws
    df['H'] = Hs

    return df
```

```python
df_Vid_1 = txt_to_df(1)
df_Vid_2 = txt_to_df(2)
df_Vid_3 = txt_to_df(3)
df_Vid_4 = txt_to_df(4)
df_Vid_5 = txt_to_df(5)
df_Vid_6 = txt_to_df(6)
```

```python
# this function make the label files correspond to each available frame
def create_labels_files(df_Vid , Vid_res , Vid_num):
    frame_numbers = df_Vid['frame_number'].unique()
    W_of_vid , H_of_vid = Vid_res
    for fn in frame_numbers:
        df_fn_selected = df_Vid.loc[df_Vid['frame_number'] == fn ]
        selected_values = df_fn_selected.values
        file_txt = open(f'labels_{Vid_num}/Vid_{Vid_num}_frame_{fn}.txt', 'w')
        for row in selected_values:
            ###
            x = int(row[1])
            y = int(row[2])
            x_center = (x + int(row[3]) / 2) / W_of_vid
            y_center = (y  + int(row[4]) / 2) / H_of_vid
            width = int(row[3]) / W_of_vid
            height = int(row[4]) / H_of_vid
            text_write = "0" +" "+ str(x_center)  +" "+  str(y_center) +" "+ str(width) +" "+ str(height) + "\n"
            ###
            file_txt.writelines(text_write)
        file_txt.close()
```

```python
from tqdm import tqdm
```

```python
#make image filder
```

```python
def create_images_file(df_Vid ,Vid_num):
    frame_numbers = df_Vid['frame_number'].unique().astype(int)
    images_folder_path = '/content/Dataset/images'
    input_video = f'/content/drive/MyDrive/Synopsis dataset/Vid{Vid_num}.mp4'
    cap = cv2.VideoCapture(input_video)  # input video
    total_number_of_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    for i in tqdm(range(1,total_number_of_frames +1)):
        success , image = cap.read()
        if not success:
            break
        if i in frame_numbers:
         # save the frame in the folder
         cv2.imwrite(images_folder_path + f'/Vid_{Vid_num}_frame_{i}.jpg' , image)
```

```python
create_images_file(df_Vid_1 , 1)
print('Done1!')
create_images_file(df_Vid_2 , 2)
print('Done2!')
create_images_file(df_Vid_3 , 3)
print('Done3!')
create_images_file(df_Vid_4 , 4)
print('Done4!')
create_images_file(df_Vid_5 , 5)
print('Done5!')
create_images_file(df_Vid_6 , 6)
print('Done6!')
```
